## 2-й эксперимент

(2-1)

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("data/exp2-1.csv", delimiter=",", skiprows=1)

I = data[:, 0]
nu = data[:, 1]

# аппроксимация прямой
a, b = np.polyfit(I, nu, 1)

# значения для линии
I_fit = np.linspace(I.min(), I.max(), 200)
nu_fit = a * I_fit + b

plt.title("2-й эксперимент (2-1)")
plt.plot(-a * I + b, nu, marker='o', linestyle='')
plt.plot(-a * I_fit + b, nu_fit, '-', label=f"$\\nu = {a:.3f} I + {b:.3f}$")
plt.xlabel(r"$\nu_I$, кГц")
plt.ylabel(r"$\nu$, kHz")


plt.show()
```
